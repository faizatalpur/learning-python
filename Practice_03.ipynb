```python
# python data types
string = "faiza"
print(string)
#numeric types
int
a = 10
b = 20
print("addition",a+b)
print("subtraction",a-b)
print("multiplication",a*b)
print("division",a/b)

#complex numbers
a=  4+4j
b=  4+2j
print(a,b)
print("addition 4+4=8,4+2=6",a+b)
print(a-b)
print(a*b)
print(a/b)

#sequence type

#list

#(it stores an ordered of items based on how they are added,it can be changeable or mutable after creation,it can containn duplicate values)
#list created by using square bracket []

f=["peach","apple","banana"]
g=[40,98,50,78]
h=[30,"banana",45,"grapes"]
i=("peach")
print(f)
print(g)
print(h)
print(type(f))
print(type(g))
print(type(h))
print(type(i))
print(f[2])
print(g[0])
print(h[1])
print(i[0])

#tuple
#it can store multiple items, it is unchangeable or immutable.
j=(20,"apples",40,"peach",30,"juices")
k=["my list",20,"apples",40,"peach"]
print(j)
print(type(j))
print(j[3])
print(k)
print(k[0])
print(type(k))

#range
#create sequence of numbers 0-9
l= range(6)
m=range(3,4,9)
print(type(l))
print(list(l))
print(m)
print(type(m))
```
