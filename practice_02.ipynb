# python operators

A Python operator is a symbol that performs an operation on one or more operands. Operands are the values that the operator acts upon.

For example, in the expression `5 + 3`, `+` is the operator, and `5` and `3` are the operands. The operator `+` performs the addition operation on the operands `5` and `3`.

Python has various types of operators:

*   **Arithmetic operators:** Perform mathematical operations like addition, subtraction, multiplication, division, etc.
*   **Comparison operators:** Compare two values and return a boolean result (True or False).
*   **Logical operators:** Combine boolean expressions.
*   **Assignment operators:** Assign values to variables.
*   **Bitwise operators:** Perform operations on individual bits.
*   **Membership operators:** Check if a value is present in a sequence.
*   **Identity operators:** Compare the memory locations of two objects.

We can explore some of these operators with code examples in the following cells.

# Comparison Operators

Comparison operators are used to compare two values and return a boolean result (`True` or `False`).

```python
# Comparison Operators

x = 10
y = 12

print("x is less than y is", x < y)
print("x < y is", x < y)
print("x == y is", x == y)
print("x != y is", x != y)
print("x >= y is", x >= y)
print("x <= y is", x <= y)
```

```python
# Arithmetic Operators

# Addition
x = 10
y = 5
print("x + y =", x + y)

# Subtraction
print("x - y =", x - y)

# Multiplication
print("x * y =", x * y)

# Division
print("x / y =", x / y)

# Modulus
print("x % y =", x % y)

# Exponentiation
print("x ** y =", x ** y)

# Floor Division
print("x // y =", x // y)
```
